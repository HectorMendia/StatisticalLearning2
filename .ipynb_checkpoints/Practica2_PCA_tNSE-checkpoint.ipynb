```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from sklearn.manifold import TSNE
from mpl_toolkits import mplot3d 
```

#### Carga de datos

```python
fashion_mnist = keras.datasets.fashion_mnist
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
plt.imshow(train_images[0], cmap = 'gray')
plt.colorbar()
```

Datos a transformar

```python
print(train_images.shape)
print(train_images[0].shape)
```

```python
image_size = train_images[0].shape[0]
largo = len(train_images)
```

```python
v_train_images = train_images.reshape(largo,image_size*image_size)
```

```python
print(v_train_images.shape)
print(v_train_images[0].shape)
```

Normalizacion de cada imagen

```python
n_train_images = (v_train_images - np.mean(v_train_images,axis=1)[:,None]) / np.std(v_train_images,axis=1)[:,None]
```

```python
print(np.std(n_train_images[0]))
print(np.mean(n_train_images[0]))
```

```python
# convresion a matriz para calculo de covarianza
#n_train_images = n_train_images.reshape(largo,image_size,image_size)
n_train_images_T = n_train_images.T
n_train_images_T.shape
```

Matriz de covarianza

```python
sigma = np.cov(n_train_images_T)
```

Eigenvalues

```python
w_train, U =  np.linalg.eig(sigma)
```

```python
w_train.shape
#v_train.shape
```

w_train se usa para saber cuantas culumnas son necesarias para representar un porcentaje

100% = sum(w_train)

```python
w_train[0:3] / sum(w_train)
```

#### Varianza preservada

*41%* lo cual es bajo

```python
sum(w_train[0:3] / sum(w_train))
```

Seleccion de k columnas

```python
U_reducida = U[0:3]
U_reducida
```

Reduccion de componentes

```python
Z = np.dot(n_train_images,U_reducida.T)
```

```python
Z.shape
```

```python
Z[0:10]
```

##### Representacion grafica

```python
# Representacion 2D
#plt.scatter(Z[:, 0], Z[:, 1])
#plt.xlabel('component 1')
#plt.ylabel('component 2')
```

```python
fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 
ax.scatter3D(Z[:, 0], Z[:, 1], Z[:, 2], color = "green"); 
plt.title("3 componentes") 
# show plot 
plt.show() 
```

```python
fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 
ax.scatter3D(Z[:, 0], Z[:, 1], Z[:, 2], c = train_labels, alpha=0.5); 
plt.title("3 componentes - PCA") 
plt.show() 
```

## t-SNE

```python
x_embed = TSNE(n_components = 3).fit_transform(n_train_images)
```

```python
x_embed.shape
```

```python
x_embed[0:10]
```

```python
fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 
ax.scatter3D(x_embed[:, 0], x_embed[:, 1], x_embed[:, 2], c = train_labels, alpha=0.5); 
plt.title("3 componentes - t-SNE") 
plt.show() 
```

### Comparacion entre ambos metodos

Debido a la cantidad de valores t-SNE es significativamente mas tardado que PCA; por lo que no es recomendable para casos donde la cantidad de componentes a reducir fuera mas grande.  O que se requiera de otros procesos o librerias para complementar esta tarea como por ejemplo procesamiento en paralelo o distribuido.  Tambien se podria considerar una combinacion entre ambos proceso de manera que PCA reduzca la primer parte y que esta sea la entrada para t-SNE aprovechando las bondades de ambos metodos.

Con estos resultados es dificil saber si alguno es mejor a otro; por lo que es necesario evaluar su *perfomance*; Este es metodo incial de un proceso mas complejo para procesar a partir de la reduccion de componentes.

### Clusterin sobre la representacion reducida

```python
from sklearn.cluster import KMeans
```

Para los datos de PCA

```python
# se decidio utilizar ball_tree debido a la cantidad de elementos que se poseen
nn_pca = KMeans(n_clusters=10, random_state=0).fit(Z)
```

```python
print("Cantidad de clustres: {}".format(nn_pca.cluster_centers_.shape))
print("Cantidad de etiquetas: {}".format(nn_pca.labels_.shape))
```

```python
knn_pca = nn_pca.cluster_centers_
knn_pca
```

```python
knn_labels_pca = nn_pca.labels_
knn_labels_pca[0:5]
```

```python
fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 
ax.scatter3D(Z[:, 0], Z[:, 1], Z[:, 2], c = knn_labels_pca, alpha=0.25); 
ax.scatter3D(knn_pca[:, 0], knn_pca[:, 1], knn_pca[:, 2], c='red', s=300, alpha=1);
plt.title("3 componentes - PCA - knn") 
plt.show() 
```

```python
fig = plt.figure(figsize = (10, 7)) 
ax = plt.axes(projection ="3d") 
ax.scatter3D(knn_pca[:, 0], knn_pca[:, 1], knn_pca[:, 2], c='red', s=200, alpha=1);
plt.title("3 componentes") 
plt.show() 
```
